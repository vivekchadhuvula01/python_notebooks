```python
'''
Matplotlib is a low-level graph plotting library in python that serves as a visualization utility.
Matplotlib was craeted by John D. Hunter.
Matplotlib is open-source and we can use it freely.

Scatter Plot
Step Plot
Bar Chart
Fill Between
Time Series 
Box Plot
Histogram

'''
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd



# make data
x = np.linspace(0, 10, 100)

y = 4 + 2 * np.sin(2 * x)

x2 = np.linspace(0, 10, 25)

y2 = 4 + 1 * np.sin(2 * x2)


# # plot
fig, ax = plt.subplots()

ax.plot(x2, y2 + 2.5, '*-', markeredgewidth=2)
ax.plot(x, y, linewidth=2.0)
ax.plot(x2, y2 - 2.5, 'o-', linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8), ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
'''
STACK PLOT
'''
days = [1,2,3,4,5,6,7]
week1=np.random.randint(10,50,7)
week2=np.random.randint(10,60,7)
week3=np.random.randint(10,70,7)

plt.stackplot(days,day1,day2,day3,colors=['pink','purple','blue'],labels=['week1','week2','week3'])
plt.legend()
plt.show()
```

```python
'''
STEP PLOT
'''
plt.step(days,day1,'violet',where='post')
plt.show()
```
